```python
codelist = ['\r', '\n', '\t']  
```

```python
stoplist = nltk.corpus.stopwords
```

```python
def text_parse(string):
    # replace non-alphanumeric with space 
    temp_string = re.sub('[^a-zA-Z]', '  ', string)    
    # replace codes with space
    for i in range(len(codelist)):
        stopstring = ' ' + codelist[i] + '  '
        temp_string = re.sub(stopstring, '  ', temp_string)      
    # replace single-character words with space
    temp_string = re.sub('\s.\s', ' ', temp_string)   
    # convert uppercase to lowercase
    temp_string = temp_string.lower()    
    # replace selected character strings/stop-words with space
    # replace multiple blank characters with one blank character
    temp_string = re.sub('\s+', ' ', temp_string)    
    return(temp_string)  
```

```python
#importing required nltk libraries
import nltk
from nltk import sent_tokenize
```

```python
# Preprocessing - used negwords from the preprocessing in HW1 which used all lowercases
# Separate the text into sentences first
scotchsplit = nltk.sent_tokenize(scotchreviews)
print(scotchsplit[:10])
```

```python
# Apply the word tokenizer to each sentence
scotchtokentext = [nltk.word_tokenize(sent) for sent in scotchsplit]
print(scotchtokentext[:2])
```

```python
scotchreviews1 = [w.lower() for w in scotchreviews]
print(len(scotchreviews1))
print(scotchreviews1[:110])
```

```python
#the output is a list of strings that contains the sentences
type(scotchtokentext)
len(scotchtokentext)
```

```python
# POS Tagging, to retrieve adjective (JJs) and adverb (RBs) tags
   
# use the Stanford POS tagger to POS tag tokens of each sentence
# this is the default tagger in nltk
scotchtaggedtext = [nltk.pos_tag(tokens) for tokens in scotchtokentext]
print(scotchtaggedtext[:2])
```

```python
# Using CHUNKING to parse sentences 
# to look for \"adjective phrases\", i.e. phrases (or chunks) that have adverbs and adjectives ('RB'+'JJ')
# First step: writing a grammar that defines the POS in the chunk
# we name this grammar \"ADJPH\" (\"ADJective PHrase\") using regexes 
    
import re
scotchgrammar_adjph = "ADJPH: {<RB.?>+<JJ.?>}"
# This regex reads as: \"find groups (\"< >\") of RBs (adverbs) together with groups of JJs (adjectives), with groups defined as
# RBs with any ending (the \".\" is a placeholder or wildcard for the \"R\" and the \"S\" at the end of RBR and RBS, 
# while \"?\" indicates \"optional character\" so RB can be found alone as well). Same regex operators apply to JJs.
```

```python
# Second step: import the nltk parser to process each sentence
scotchchunk_parser_adj = nltk.RegexpParser(scotchgrammar_adjph)
   
scotchadjph_tags = []
for sent in scotchtaggedtext:
    if len(sent) > 0:
     tree = scotchchunk_parser_adj.parse(sent)
     for subtree in tree.subtrees():
       if subtree.label() == 'ADJPH':
         scotchadjph_tags.append(subtree)
```

```python
# Visualizing the actual adjective phrase
scotchadjective_phrases = []
for sent in scotchadjph_tags:
    temp = ''
    for w, t in sent:
      temp += w+ ' ' 
      scotchadjective_phrases.append(temp)
print('First 10 adjective phrases: ', scotchadjective_phrases[:10])
```

```python
# FREQUENCY DISTRIBUTIONS
# Top 50 adjective phrases
scotchfreq_adjph = nltk.FreqDist(scotchadjective_phrases)
print('Top adjective phrases by frequency: ')
for word, freq in scotchfreq_adjph.most_common(50):
    print(word, freq)
```

```python
#print the list of our sentences:
print('Length of adjective phrase sentences: ', len(scotchadjph_tags))
```

```python
# Now we look for \"adverb phrases\" or chunks that have 2 consecutive adverbs ('RB')
# First step: writing a grammar that defines POS rules of the adverb phrase the chunk
# we name this grammar \"ADVPH\" (\"ADVerb PHrase\")
scotchgrammar_advph = "ADVPH: {<RB>+<RB>}"
```

```python
# Second step import the nltk parser to process each sentence
scotchchunk_parser_adv = nltk.RegexpParser(scotchgrammar_advph)

scotchadvph_tags = []
for sent in scotchtaggedtext:
    if len(sent) > 0:
      tree = scotchchunk_parser_adv.parse(sent)
      for subtree in tree.subtrees():
        if subtree.label() == 'ADVPH':
          scotchadvph_tags.append(subtree)
```

```python
# Visualizing the actual adjective phrase
scotchadverb_phrases = []
for sent in scotchadvph_tags:
    temp = ''
    for w, t in sent:
      temp += w+ ' '
      scotchadverb_phrases.append(temp)
print('First 10 adverb phrases: ', scotchadverb_phrases[:10])
```

```python
# top 50 adverb phrases
scotchfreq_advph = nltk.FreqDist(scotchadverb_phrases)

print('Top adverb phrases by frequency: ')
for word, freq in scotchfreq_advph.most_common(50):
    print(word, freq)
```

```python
#print the list of our sentences:
print('Length of adverb phrase sentences: ', len(scotchadvph_tags))
```

```python
# Top 50 adjective tokens

scotchadjective_tokens = []
for sentence in scotchtaggedtext:
    for word, pos in sentence:
      if pos in ['JJ', 'JJR', 'JJS']: # adjective, comparative, superlative
        if len(word)>1:
          scotchadjective_tokens.append(word)
          scotchfreq_adjective = nltk.FreqDist(scotchadjective_tokens)
    
for word, freq in scotchfreq_adjective.most_common(50):
    print(word,freq)
```

```python
# Top 50 adverb tokens
    
scotchadverb_tokens = []
for sentence in scotchtaggedtext:
    for word, pos in sentence:
      if pos in ['RB', 'RBR', 'RBS']: # adverb, comparative, superlative
        if len(word)>1:
          scotchadverb_tokens.append(word)
          scotchfreq_adverb = nltk.FreqDist(scotchadverb_tokens)
    
for word, freq in scotchfreq_adverb.most_common(50):
  print(word,freq)
```

```python
# Now we have two lists of POS tags combinations we can compare
# We need to get the sentences back from the tagging exercise and run some stats
    
# Create a list of original sentences from the ADJECTIVE phrase subset:
scotchadjph_whole_sentences = []
```

```python
# loop over the sentences in the adjective phrase sentences we created:
for sents in scotchadjph_tags:
    temp=''
    for (word,tag) in sents:
      temp += word+' '
      scotchadjph_whole_sentences.append(temp)    
print(len(scotchadjph_whole_sentences))
```

```python
# Create a list of original sentences from the ADVERB phrase subset:
scotchadvph_whole_sentences = []
```

```python
# loop over the sentences in the adverb phrase sentences we created:
for sents in scotchadvph_tags:
    temp=''
    for (word,tag) in sents:
      temp += word+' '
      scotchadvph_whole_sentences.append(temp)
print(len(scotchadvph_whole_sentences))
```

```python
# OPTIONAL STEP: Combine lists together to have a single list of adjective/adverb phrases:
# Useful to know which sentences are heavy in qualifiers
   
# create a new variable to store all adjective phrase sentences
scotchadv_adj_phrase_sentences = scotchadjph_whole_sentences
```

```python
# iterate over adverb phrase sentences
for sent in scotchadvph_whole_sentences:
    # if a sentence is not in the adjective phrases list imported
    if sent not in scotchadv_adj_phrase_sentences:
    # attach that sentence
      scotchadv_adj_phrase_sentences.append(sent)
# print the length of the list (i.e. number of sentences with both adjective and adverb phrases)
print(len(scotchadv_adj_phrase_sentences))
```

```python
# CORPUS STATISTICS--SENTENCES LENGTH

# Calculating the average length of sentences in the entire corpus
scotchtotal_corpus = sum(len(sent) for sent in scotchsplit) # remember: 'textsplit' is our text split into sentences
print(scotchtotal_corpus / len(scotchsplit))
```

```python
# Calculate the average length of an adjective phrase sentence
# We can then compare the average length of the adjective phrases to 
# the average sentences we calculated for all sentences 

# adjph_whole_sentences stores our adjective phrases
scotchtotal_adjph_sentences = sum(len(sent) for sent in scotchadjph_whole_sentences) 
print(scotchtotal_adjph_sentences / len(scotchadjph_whole_sentences))
```

```python
# Calculate the average length of an adverb phrase sentence
# We can then compare the average length of the adverb phrases to 
# the average sentences we calculated for all sentences 

# advph_whole_sentences stores our adverb phrases
scotchtotal_advph_sentences = sum(len(sent) for sent in scotchadvph_whole_sentences) 
print(scotchtotal_advph_sentences / len(scotchadvph_whole_sentences))
```
